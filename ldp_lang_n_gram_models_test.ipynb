```python
def n_gram_model(utt_list): 
    
    # Create a placeholder for model
    model = defaultdict(lambda: defaultdict(lambda: 0))

    # Count frequency of co-occurance  
    for sentence in utt_list:
        for w1, w2 in bigrams(sentence, pad_right=True, pad_left=True):
            model[(w1)][w2] += 1

    #Lets transform the counts to probabilities
    for w1_ in model:
        total_count = float(sum(model[w1_].values()))
        for w2 in model[w1_]:
            model[w1_][w2] /= total_count
    
    
    return model
```

```python
def det_noun_probs(det, model):
    td ={k: v for k, v in sorted(dict(model[det]).items(), key=lambda item: item[1], reverse=True)}

    ten_probs = list(td.keys())[:10]
    
    return [td, ten_probs]
```

```python
def viz_det_counts(ptd, p_ten, ctd, c_ten, sub, session, det, speaker):
    total_count = 0
#     print(ptd)
    print(ctd)
    ptd = list(ptd.keys())
     
    ctd = list(ctd.keys())
    
    for item in ctd:
        if item in ptd:
            total_count += 1
            
    ten_count = 0
    
    for i in c_ten:
        if i in p_ten:
            ten_count += 1
            
    di = {'shared_top_ten_items': ten_count, 'all_items': total_count}
            
    pd.DataFrame(di, index=[0]).plot(kind='bar', title=f'{speaker}_{sub}_{session}_det_{det}_shared_nouns_counts')

    fig1 = plt.gcf()
    plt.show()
    plt.draw()
    fig1.savefig(f'{sub}_{session}_{speaker}_det_{det}_shared_nouns_counts.png', dpi=100)
    plt.show()
```

```python
def viz_det_props(ptd, p_ten, ctd, c_ten, sub, session, det, speaker):
    total_count = 0
#     print(ptd)
    print(ctd)
    ptd = list(ptd.keys())
     
    ctd = list(ctd.keys())
    
    for item in ctd:
        if item in ptd:
            total_count += 1
            
    ten_count = 0
    
    for i in c_ten:
        if i in p_ten:
            ten_count += 1
            
    di = {'shared_top_ten_items': (ten_count/10), 'all_items': 2*(total_count/(len(ctd) + len(ptd)))}
            
    pd.DataFrame(di, index=[0]).plot(kind='bar', title=f'{speaker}_{sub}_{session}_det_{det}_shared_nouns_props')

    fig1 = plt.gcf()
    plt.show()
    plt.draw()
    fig1.savefig(f'{sub}_{session}_{speaker}_det_{det}_shared_nouns_props.png', dpi=100)
    plt.show()
```

```python
def viz_det_noun_probs(td, sub, session, det, speaker):    
    
    pd.DataFrame(td, index=[0]).iloc[:, :10].plot(kind='bar', title=f'{speaker}_{sub}_{session}_det_{det}_nps_probs')

    fig1 = plt.gcf()
    plt.show()
    plt.draw()
    fig1.savefig(f'{sub}_{session}_{speaker}_det_{det}_nps_probs.png', dpi=100)
    plt.show()
```

```python
tsesh = [1,2,3,4]
for sub in subj_ids:
    for sesh in tsesh:
        c_utts = mdf[(mdf['subject'] ==sub) & (mdf['session'] == sesh)]['c_utts_orig'].tolist()
        p_utts = mdf[(mdf['subject'] ==sub) & (mdf['session'] == sesh)]['p_utts_orig'].tolist()
        p_model = n_gram_model(p_utts)
        c_model = n_gram_model(c_utts)
        p_det_nps_probs = det_noun_probs('a', p_model)
        c_det_nps_probs = det_noun_probs('a', c_model)
        try:
            viz_det_props(p_det_nps_probs[0], p_det_nps_probs[1],
                                   c_det_nps_probs[0], c_det_nps_probs[1],
                                   sub, sesh, 'a', 'parent_child_props')
        except:
            print(sub, sesh)
        
        
```
